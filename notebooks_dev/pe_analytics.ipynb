#### Quantidade de votos para Prefeito em Recife por turno

O código abaixo realiza a filtragem para exibição do número total de votos para Prefeito na cidade do Recife por turno

```python
df_pref_rec = df_recife[df_recife['DS_CARGO'] == 'Prefeito']
pref_rec_turn_1 = df_pref_rec[df_pref_rec['NR_TURNO'] == 1]
total_votos_pref_rec_turn_1 = pref_rec_turn_1['QT_VOTOS'].sum()
print("Quantidade Total de Votos para Prefeito em Recife no PRIMEIRO turno: ", total_votos_pref_rec_turn_1)

pref_rec_turn_2 = df_pref_rec[df_pref_rec['NR_TURNO'] == 2]
total_votos_pref_rec_turn_2 = pref_rec_turn_2['QT_VOTOS'].sum()
print("Quantidade Total de Votos para Prefeito em Recife no SEGUNDO turno: ", total_votos_pref_rec_turn_2)
```

Criação do grafico para exibição do resultado acima através da biblioteca matplotlib

```python
# Totais de votos para o primeiro e segundo turnos
total_votos_pref_rec_turn_1 = 927167
total_votos_pref_rec_turn_2 = 911314

# Definição dos turnos e dos votos correspondentes
turnos = ['Primeiro Turno', 'Segundo Turno']
votos_por_turno = [total_votos_pref_rec_turn_1, total_votos_pref_rec_turn_2]

# Ajuste da largura das barras
largura_das_barras = 0.5

# Plotar o gráfico de barras com largura ajustada
plt.figure(figsize=(6, 4))

# Criação do gráfico de barras
barras = plt.bar(turnos, votos_por_turno, color='skyblue', width=largura_das_barras)

# Adicionar rótulos e título
plt.xlabel('Turno')
plt.ylabel('Total de Votos')
plt.title('Total de Votos para Prefeito em Recife por Turno (Gráfico de Barras)')

# Adicionar os números de votos sobre cada barra
for barra, votos in zip(barras, votos_por_turno):
    plt.text(barra.get_x() + barra.get_width() / 2, 
             barra.get_height() + 10000,  # Ajuste vertical para a posição do texto
             f'{votos:,}',  # Formatação para adicionar vírgulas nos milhares
             ha='center', va='bottom', fontsize=10)

# Exibir o gráfico
plt.show()
```

#### Candidatos mais votados a Prefeito do Recife no 1º turno


Exibição dos 5 primeiros candidatos mais votados no 1º turno na eleição para Prefeito do Recife. Foi realizada a filtragem para remoção dos votos brancos e nulos, seguido pela soma dos votos de cada candidato e definição da exibição em ordem decrescente.

```python
# Filtragem para excluir 'VOTO NULO' e 'VOTO BRANCO'
# O método isin() é usado para verificar se os valores de uma série ou coluna de um DataFrame estão presentes em uma lista de valores especificada
# Ele retorna uma série booleana que indica se cada elemento está ou não na lista de valores.
votos_validos_pref = pref_rec_turn_1[~pref_rec_turn_1['NM_VOTAVEL'].isin(['VOTO NULO', 'VOTO BRANCO'])]
soma_votos_por_candidato_pref = votos_validos_pref.groupby('NM_VOTAVEL')['QT_VOTOS'].sum()
soma_votos_por_candidato_ord_pref = soma_votos_por_candidato_pref.sort_values(ascending=False)
candidatos_mais_votados = soma_votos_por_candidato_ord_pref.head(5)

print("Ordem dos 5 Candidatos a Prefeito no Recife mais votados no PRIMEIRO turno (excluindo VOTO NULO e VOTO BRANCO): \n", candidatos_mais_votados)
```

#### Prefeitos mais votados por Município onde houve no 2º Turno

```python

df_pref_turno_2 = df[(df['DS_CARGO'] == 'Prefeito') & (df['NR_TURNO'] == 2)]
total_votos_pref_turno_2_por_municipio = df_pref_turno_2.groupby(['NM_MUNICIPIO', 'NM_VOTAVEL'])['QT_VOTOS'].sum()
candidatos_mais_votados_por_municipio = total_votos_pref_turno_2_por_municipio.groupby('NM_MUNICIPIO').idxmax()
soma_votos_por_municipio = total_votos_pref_turno_2_por_municipio.groupby('NM_MUNICIPIO').sum()
soma_votos_por_municipio_sorted = soma_votos_por_municipio.sort_values(ascending=False)

for municipio in soma_votos_por_municipio_sorted.index:
    idx = candidatos_mais_votados_por_municipio.loc[municipio]
    candidato_mais_votado = idx[1] 
    votos_candidato_mais_votado = total_votos_pref_turno_2_por_municipio.loc[(municipio, candidato_mais_votado)]
    
    print(f"Município: {municipio}")
    print(f"Candidato Mais Votado: {candidato_mais_votado}")
    print(f"Total de Votos: {votos_candidato_mais_votado}")
    print() 
```

```python
municipios = soma_votos_por_municipio_sorted.index
votos = soma_votos_por_municipio_sorted.values
candidatos_mais_votados = [candidatos_mais_votados_por_municipio.loc[municipio][1] for municipio in municipios]

plt.figure(figsize=(12, 8))
plt.barh(municipios, votos, color='orange', label='Votos do Candidato Mais Votado', alpha=0.7)

for i, municipio in enumerate(municipios):
    plt.text(votos[i] + 1000, i, f'{candidatos_mais_votados[i]} ({votos[i]:,.0f} votos)', va='center')

plt.title('Total de Votos do Candidato Mais Votado por Município (2º Turno)')
plt.xlabel('Total de Votos')
plt.ylabel('Município')
plt.legend()

plt.show()
```

Para a eleição de Vereadores adota-se o **sistema proporcional** de votação, também utilizado para eleição de Deputados Estaduais e Federais.

Para a eleição para prefeituras, governadores, Senado e Presidência adota-se o **sistema majoritário** em que aquele ou aquela que recebeu mais votos se elege.

O sistema proporcional, por outro lado, visa priorizar as legendas (os partidos político).
Ou seja, visa priorizar a agregação ideológica em detrimento de uma figura particular.

Um exemplo da diferença causada pelo sistema proporcional em relação ao majoritário foi nas eleições de 2012 para vereador no Recife, em que um candito foi o 3º mais votado e ainda assim não foi eleito.

Isso ocorre pelo **quociente eleitoral** que é o número de votos válidos para determinado cargo dividido pelo número de vagas a preencher para determinado cargo.

Por outro lado, o **quociente partidário** que define o número de vagas que cada partido tem direito, que é o resultado do número de votos válidos para determinado partido ou federação dividido pelo quociente eleitoral.
